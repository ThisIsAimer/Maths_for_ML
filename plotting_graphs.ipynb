# plotting with python

```python
import numpy
import matplotlib.pyplot as plot
```

```python
time = numpy.linspace(0,40,500)
```

### distance travelled by robber = 2.5t and the cop started 5 mins = 3(t-5)

```python
distance_robber = 2.5 * time
distance_cop = 3 * (time-5)
```

### plotting

```python
fig, ax = plot.subplots()
plot.title("A bank robber caught")
plot.xlabel("time (in mins)")
plot.ylabel("distance (in kms)")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(time,distance_robber,c="green")
ax.plot(time,distance_cop,c="red")
plot.axvline(x=30,color="purple", linestyle="--")
plot.axhline(y=75,color="purple", linestyle="--")
```
